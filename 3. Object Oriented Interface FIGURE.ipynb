## Object Oriented Interface

While it is easy to quickly generate plots with the matplotlib.pyplot module, the use of object-oriented approach is recommended as it gives more control and customization of your plots. Most of the functions are also available in the matplotlib.axes.Axes class.

The main idea behind using the more formal object-oriented method is to create figure objects and then just call methods or attributes off of that object. This approach helps better in dealing with a canvas that has multiple plots on it.

In object-oriented interface, Pyplot is used only for a few functions such as figure creation, and the user explicitly creates and keeps track of the figure and axes objects. At this level, the user uses Pyplot to create figures, and through those figures, one or more axes objects can be created. These axes objects are then used for most plotting actions.

### Working with Object Oriented Interface

To begin with, we create a figure instance which provides an empty canvas.

Now add axes to figure. The add_axes() method requires a list object of 4 elements corresponding to left, bottom, width and height of the figure. Each number must be between 0 and 1 

Adding Title and Labels to the axes

Making new dataframe with information of Pakistan only

Plot gdpPerCapita to Year

```python
print(type(plt))

fig = plt.figure()
print('Type of fig is : ' ,type(fig))

ax = fig.add_axes([0, 0, 1, 1])
print('Type of ax is : ' ,type(ax))

ax.set_title("GDP per Capita")
ax.set_xlabel("Year")
ax.set_ylabel("GDP per Capita")

pak = df[df['country'] == 'Pakistan']

ax.plot(pak['year'], pak['gdpPerCapita'])
```

__Note__:<br> We saw above __ax__ is an instance of Axes class.

### Exploring Parameters One By One

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.set_title("It is a default Figure Instance without any Parameter Defined")
ax.set_xlabel("Year")
ax.set_ylabel("GDP per Capita")

pak = df[df['country'] == 'Pakistan']

ax.plot(pak['year'], pak['gdpPerCapita'])
```

```python
fig = plt.figure(
    
    figsize=(8,5)

)

ax = fig.add_axes([0, 0, 1, 1])

ax.set_title("Set Figure Size to (8, 5)")
ax.set_xlabel("Year")
ax.set_ylabel("GDP per Capita")

pak = df[df['country'] == 'Pakistan']

ax.plot(pak['year'], pak['gdpPerCapita'])
```

```python
fig = plt.figure(
    
    figsize=(8,5),
    facecolor= '#66ed76'

)

ax = fig.add_axes([0, 0, 1, 1])

ax.set_title("Set Figure Size to (8, 5)\nFace Color : #66ed76")
ax.set_xlabel("Year")
ax.set_ylabel("GDP per Capita")

pak = df[df['country'] == 'Pakistan']

ax.plot(pak['year'], pak['gdpPerCapita'])
```

```python
fig = plt.figure(
    
    figsize=(8,5),
    facecolor= '#66ed76',
    edgecolor='#369c42',
    linewidth=5

)

ax = fig.add_axes([0, 0, 1, 1])

ax.set_title("Set Figure Size to (8, 5)\nFace Color : #66ed76\nEdgecolor : #369c42\nLine Width : 2")
ax.set_xlabel("Year")
ax.set_ylabel("GDP per Capita")

pak = df[df['country'] == 'Pakistan']

ax.plot(pak['year'], pak['gdpPerCapita'])
```

```python
fig = plt.figure(
    
    facecolor= '#66ed76',
    edgecolor='#369c42',
    linewidth=5,
    dpi=50

)

ax = fig.add_axes([0, 0, 1, 1])

ax.set_title("Face Color : #66ed76\nEdgecolor : #369c42\nLine Width : 2\ndpi : 50")
ax.set_xlabel("Year")
ax.set_ylabel("GDP per Capita")

pak = df[df['country'] == 'Pakistan']

ax.plot(pak['year'], pak['gdpPerCapita'])
```

```python
fig = plt.figure(
    
    facecolor= '#66ed76',
    edgecolor='#369c42',
    linewidth=5,
    dpi=100

)

ax = fig.add_axes([0, 0, 1, 1])

ax.set_title("Face Color : #66ed76\nEdgecolor : #369c42\nLine Width : 2\ndpi : 100")
ax.set_xlabel("Year")
ax.set_ylabel("GDP per Capita")

pak = df[df['country'] == 'Pakistan']

ax.plot(pak['year'], pak['gdpPerCapita'])
```

```python
fig = plt.figure(
    
    facecolor= '#66ed76',
    edgecolor='#369c42',
    linewidth=5

)

# rect =  [ left, bottom, width, height ]

ax = fig.add_axes([0, 0, 1, 1])

ax.set_title("Comparing GDP per Capita\nrect : [0, 0, 1, 1]")
ax.set_xlabel("Year")
ax.set_ylabel("GDP per Capita")

pak = df[df['country'] == 'Pakistan']
tur = df[df['country'] == 'Turkey']

ax.plot(pak['year'], pak['gdpPerCapita'])
ax.plot(tur['year'], tur['gdpPerCapita'])
```
